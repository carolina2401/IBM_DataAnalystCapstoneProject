#### Step 4: Analysis of Experience and Job Satisfaction


Analyze the relationship between years of professional coding experience (`YearsCodePro`) and job satisfaction (`JobSat`). Summarize `YearsCodePro` and calculate median satisfaction scores based on experience ranges.

- Create experience ranges for `YearsCodePro` (e.g., `0-5`, `5-10`, `10-20`, `>20` years).

- Calculate the median `JobSat` for each range.

- Visualize the relationship using a bar plot or similar visualization.


```python
## Write your code here

df['YearsCodePro'].value_counts()
```

```python
# Clean and convert YearsCodePro to numeric
years = pd.to_numeric(
    df['YearsCodePro'].replace({
        'Less than 1 year': 0,
        'More than 50 years': 51
    }),
    errors='coerce'
)

# Create experience ranges
df['ExperienceLevel'] = pd.cut(
    years,
    bins=[0, 5, 10, 20, 100],
    labels=['0-5 years', '5-10 years', '10-20 years', '>20 years'],
    include_lowest=True
)

df['ExperienceLevel'].value_counts()
```

```python
# Performing a histogram of JobSat to understand its distribution

plt.figure(figsize=(8,5))

plt.hist(df['JobSat'], bins=10, alpha=0.7)

plt.title('Distribution of Job Satisfaction')
plt.xlabel('Job Satisfaction Score')
plt.ylabel('Frequency')

plt.show()
```

```python
# Calculate median Job Satisfaction for each experience range
median_jobsat = df.groupby('ExperienceLevel')['JobSat'].median()

# Display summary
print(median_jobsat)
```

```python
# Visualization
plt.figure(figsize=(8,5))

median_jobsat.plot(kind='bar')

plt.title('Median Job Satisfaction by Experience Level')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Median Job Satisfaction')
plt.xticks(rotation=0)

plt.show()
```

#### Step 6: Analyzing Remote Work Preferences by Job Role


Analyze trends in remote work based on job roles. Use the `RemoteWork` and `Employment` columns to explore preferences and examine if specific job roles prefer remote work more than others.

- Use a count plot to show remote work distribution.

- Cross-tabulate remote work preferences by employment type (e.g., full-time, part-time) and job roles.


```python
## Write your code here

print(df['RemoteWork'].value_counts())
print()
print(df['Employment'].value_counts().head(10))
```

```python
## Recode the employment to make broader categories

def simplify_employment(x):
    first = x.split(';')[0]
    
    if first.startswith('Employed'):
        return 'Employed'
    elif first.startswith('Independent'):
        return 'Independent'
    elif first.startswith('Student'):
        return 'Student'
    elif 'but looking for work' in first:
        return 'Unemployed - Looking'
    elif 'and not looking for work' in first:
        return 'Unemployed - Not Looking'
    elif first.startswith('Retired'):
        return 'Retired'
    else:
        return 'Other'

df['Employment_Group'] = df['Employment'].apply(simplify_employment)
```

```python
# inspect the new column

print(df['Employment_Group'].value_counts())
```

```python
# Visualization of remote work column

plt.figure(figsize=(10,5))

sns.countplot(x=df['RemoteWork'])

plt.title('Remote Work Distribution')
plt.xlabel('Remote Work Type')
plt.ylabel('Number of Respondents')

plt.xticks(rotation=0)
plt.show()
```

```python
# Cross-tab RemoteWork vs Employment_Group

remote_emp = pd.crosstab(df['Employment_Group'], df['RemoteWork'])
remote_emp
```

```python
# Normalize for comparison (row percentages — VERY IMPORTANT insight)

remote_emp_pct = pd.crosstab(
    df['Employment_Group'],
    df['RemoteWork'],
    normalize='index'
) * 100

remote_emp_pct
```

```python
# Visualize relationship (stacked bar plot)

remote_emp_pct.plot(kind='barh', stacked=True, figsize=(12,8))

plt.title('Remote Work Preferences by Employment Type')
plt.xlabel('Percentage (%)')
plt.ylabel('Employment Type')

plt.legend(title='Remote Work', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xticks(rotation=0)

plt.show()
```

```python
# Alternative with seaborn

remote_emp_pct.plot(kind='bar', stacked=True, figsize=(9,6), colormap='viridis')

plt.title('Remote Work Preferences by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Percentage (%)')

plt.legend(
    title='Remote Work',
    bbox_to_anchor=(1.05, 1),
    loc='upper left'
)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Step 7: Analyzing Programming Language Trends by Region


Analyze the popularity of programming languages by region. Use the `LanguageHaveWorkedWith` column to investigate which languages are most used in different regions.

- Filter data by country or region.

- Visualize the top programming languages by region with a bar plot or heatmap.


```python
print("Counts programming languages:", df['LanguageHaveWorkedWith'].value_counts().head(10))
```

```python
# Count individual languages

(
    df['LanguageHaveWorkedWith']
    .dropna()
    .str.split(';')
    .explode()
    .value_counts()
)
```

```python
# Unique individual languages

(
    df['LanguageHaveWorkedWith']
    .dropna()
    .str.split(';')
    .explode()
    .nunique()
)
```

```python
# Check number of unique countries

print("Number of unique countries:", df['Country'].nunique())
print()
print("Counts of each country values:", df['Country'].value_counts())
```

```python
# Split and individualize languages


lang_df = (
    df[['Country', 'LanguageHaveWorkedWith']]
    .dropna()
)

lang_df['LanguageHaveWorkedWith'] = (
    lang_df['LanguageHaveWorkedWith']
    .str.split(';')
)

lang_df = lang_df.explode('LanguageHaveWorkedWith')

lang_df.head()
```

```python
# Next visualization steps for selected countries

selected_countries = [
    'United States of America',
    'Germany',
    'India',
    'United Kingdom of Great Britain and Northern Ireland',
    'Ukraine'
]

lang_filtered = lang_df[
    lang_df['Country'].isin(selected_countries)
]

lang_filtered.head()
```

```python
# Counting languages by country

language_counts = (
    lang_filtered
    .groupby(['Country', 'LanguageHaveWorkedWith'])
    .size()
    .reset_index(name='Count')
)

language_counts.head()
```

```python
# Selecting top 10 languages overall

top_languages = (
    language_counts
    .groupby('LanguageHaveWorkedWith')['Count']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .index
)

top_languages
```

```python
# # Keep only top languages

language_counts_top = language_counts[
    language_counts['LanguageHaveWorkedWith'].isin(top_languages)
]

language_counts_top.head()
```

```python
# Bar plot

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14,8))

sns.barplot(
    data=language_counts_top,
    x='LanguageHaveWorkedWith',
    y='Count',
    hue='Country'
)

plt.title('Top Programming Languages by Country')

plt.xlabel('Programming Language')
plt.ylabel('Count')

plt.xticks(rotation=45)

plt.legend(
    title='Country',
    bbox_to_anchor=(1.05, 1),
    loc='upper left'
)

plt.show()
```

```python
# Heatmap table

heatmap_data = language_counts_top.pivot_table(
    index='Country',
    columns='LanguageHaveWorkedWith',
    values='Count',
    fill_value=0
)

heatmap_data
```

```python
# 8. Heatmap visualization

plt.figure(figsize=(12,6))

sns.heatmap(
    heatmap_data,
    annot=True,
    fmt='.0f',
    cmap='Blues'
)

plt.title('Programming Language Popularity by Country')

plt.xlabel('Programming Language')
plt.ylabel('Country')

plt.show()
```

#### Step 9: Educational Background and Employment Type


Explore how educational background (`EdLevel`) relates to employment type (`Employment`). Use cross-tabulation and visualizations to understand if higher education correlates with specific employment types.


```python
## Write your code here

print("Counts Education Level:", df['EdLevel'].value_counts())
```

```python
# Creating a new column with broader categories for educational level [a similar code was used before for employment  variable]

def simplify_education(x):
    if pd.isna(x):
        return 'Other'
    
    if 'Primary' in x or 'Secondary' in x:
        return 'High school or lower'
    
    elif 'Some college' in x or 'without earning a degree' in x:
        return 'Some university (no degree)'
    
    elif 'Associate' in x:
        return 'Associate degree'
    
    elif 'Bachelor' in x:
        return 'Bachelor degree'
    
    elif 'Master' in x:
        return 'Master degree'
    
    elif 'Professional degree' in x or 'Ph.D' in x or 'doctoral' in x.lower():
        return 'Doctoral / Professional'
    
    else:
        return 'Other'


df['Education_Group'] = df['EdLevel'].apply(simplify_education)
```

```python
# Checking the new variable

df['Education_Group'].value_counts()
```

```python
# Cross-tab (Education vs Employment)

edu_emp = pd.crosstab(
    df['Education_Group'],
    df['Employment_Group']
)

edu_emp
```

```python
# get percentages for the crosstab

edu_emp_pct = pd.crosstab(
    df['Education_Group'],
    df['Employment_Group'],
    normalize='index'
) * 100

edu_emp_pct.round(1)
```

```python
# Visualization with a heatmap

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))

sns.heatmap(
    edu_emp_pct,
    annot=True,
    fmt=".1f",
    cmap="Blues"
)

plt.title("Education Level vs Employment Type (%)")
plt.xlabel("Employment Type")
plt.ylabel("Education Level")

plt.show()
```
